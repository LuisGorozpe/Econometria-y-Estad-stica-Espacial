```r
xy <- expand.grid(x = seq(0, 100, length.out = 30), 
                  y = seq(0, 100, length.out = 30))
model_truth <- vgm(psill = 50, model = "Sph", range = 30, nugget = 10)
g_dummy <- gstat(formula = z ~ 1, locations = ~x+y, dummy = TRUE, 
                 beta = 20, model = model_truth)
yy <- predict(g_dummy, newdata = xy, nsim = 1)
sf_data <- st_as_sf(yy, coords = c("x", "y"))

# 2. Empirical Variogram
v_emp <- variogram(sim1 ~ 1, data = sf_data)

# 3. Fit Theoretical Model
# We fit a Spherical model to the empirical points
v_fit <- fit.variogram(v_emp, vgm(psill=40, model="Sph", range=25, nugget=5))
```

```r
summary(v_fit)
```

```r
plot(v_fit, cutoff = 50 ,main = "Example")
```
